#### How to create dynamic prompts

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
```

```python
prompts = [
{"role": "system", "content": "You are a helpful assistant that translates English to French."},
{"role": "user", "content": "Translate the following English text to French: 'Hello, how are you?'."}
] 

response = model.invoke(prompts)
print(response)
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")

def transform(result: str) -> str:
    return result.upper()
out = StrOutputParser()
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a helpful assistant that translates English to {language}."),
        ("user", "Translate the following English text to {language}: '{text}'")
    ]
)

# response = model.invoke(prompt.invoke({"text":"Hello, how are you?", "language":"French"}))
# print(response)
```

```python
chain = prompt | model | out | transform
```

```python
res = chain.invoke({"text":"Hello, how are you?", "language":"Hinglish"})
print(res)
```
