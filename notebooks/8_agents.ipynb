```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_core.tools import tool
```

```python
@tool
def add_numbers(a: int, b: int) -> int:
    """Add two numbers together."""
    return a + b


@tool
def multiply_numbers(a: int, b: int) -> int:
    """Multiply two numbers together."""
    return a * b
```

```python
# add_numbers.invoke({"a": 5, "b": 7})
```

```python
from langchain_ollama import ChatOllama
from langchain.agents import create_agent
```

```python
model = ChatOllama(model="qwen3:1.7b")
```

```python
agent = create_agent(
  model=model, 
  tools=[add_numbers, multiply_numbers],
  system_prompt= "you are a junior school teacher who teaches math and always uses the tools to answer the question. If the question is not related to math, you can just answer it without using the tools."
  )
```

```python
response = agent.invoke({"messages": [{"role": "user", "content": "what is 5 + 7 * 100?"}]})
```
